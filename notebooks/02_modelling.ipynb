## Cross validation

### Run cross validation

```python
from sklearn.model_selection import cross_val_score

def cv_rmse(model, X, y, cv=5):
    scores = cross_val_score(model, X, y, 
                             scoring="neg_root_mean_squared_error", 
                             cv=cv, n_jobs=-1)
    scores = -scores  # flip sign (sklearn returns negative)
    print(f"RMSE scores: {scores.round(4)}")
    print(f"Mean: {scores.mean():.4f} | Std: {scores.std():.4f}")
    return scores
```

CV scores

## Hyperparameter tuning

### Tune XGBoost and LightGBM with Optuna

```python
import optuna
from sklearn.model_selection import cross_val_score
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor

optuna.logging.set_verbosity(optuna.logging.WARNING)  # suppress noise

def tune_model(model_type, X, y, n_trials=50):

    def objective(trial):

        if model_type == "xgboost":
            params = {
                "n_estimators": trial.suggest_int("n_estimators", 100, 1000),
                "learning_rate": trial.suggest_float("learning_rate", 0.01, 0.3, log=True),
                "max_depth": trial.suggest_int("max_depth", 3, 8),
                "subsample": trial.suggest_float("subsample", 0.6, 1.0),
                "colsample_bytree": trial.suggest_float("colsample_bytree", 0.6, 1.0),
            }
            model = XGBRegressor(**params, random_state=42, n_jobs=-1)

        elif model_type == "lightgbm":
            params = {
                "n_estimators": trial.suggest_int("n_estimators", 100, 1000),
                "learning_rate": trial.suggest_float("learning_rate", 0.01, 0.3, log=True),
                "num_leaves": trial.suggest_int("num_leaves", 20, 100),
                "subsample": trial.suggest_float("subsample", 0.6, 1.0),
                "colsample_bytree": trial.suggest_float("colsample_bytree", 0.6, 1.0),
            }
            model = LGBMRegressor(**params, random_state=42, verbose=-1)

        # Wrap in full pipeline
        pipeline = Pipeline([
            ("features", FeatureEngineer()),
            ("preprocessing", build_preprocessor(num_cols, cat_cols)),
            ("model", model)
        ])

        scores = cross_val_score(pipeline, X, y,
                                 scoring="neg_root_mean_squared_error",
                                 cv=5, n_jobs=-1)
        return -scores.mean()

    study = optuna.create_study(direction="minimize")
    study.optimize(objective, n_trials=n_trials)

    print(f"Best RMSE: {study.best_value:.4f}")
    print(f"Best params: {study.best_params}")
    return study.best_params
```

### Run tuning

```python
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
from sklearn.pipeline import Pipeline
from src.features import FeatureEngineer
from src.preprocessing import build_preprocessor

print("Tuning XGBoost...")
best_xg_params = tune_model("xgboost", X, y, n_trials=50)

print("\nTuning LightGBM...")
best_lgbm_params = tune_model("lightgbm", X, y, n_trials=50)
```
